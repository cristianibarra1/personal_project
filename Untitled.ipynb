```python
'''category graph that show the percent of all category'''
labels = pd.concat([df5.subject.value_counts(),df5.subject.value_counts(normalize=True)], axis=1)
labels.columns = ['n', 'percent']
plt.figure(figsize=(16,16))
mylabels = df5['subject']
textprops = {"fontsize":12}
textprops = {"fontsize":12}
plt.pie(labels.percent, labels = labels.index, textprops=textprops, autopct='%.1f%%')
plt.legend(loc='upper left')
plt.title('Overall Distribution',fontsize=18)
plt.show() 
```

# Does subject effect interation of subcribers?

```python
plt.figure(figsize=(18,8))
sns.violinplot(data=df5,x='num_subscribers',y='level')
```

```python
plt.figure(figsize=(18,8))
plt.title('Subcribers by subject')
sns.barplot(data=df5,x='num_subscribers',y='subject',hue='level')
plt.show()
```

# Does price effect the subject activity?

```python
plt.figure(figsize=(18,8))
plt.title('Subcribers by subject')
sns.barplot(data=df5,x='num_subscribers',y='subject',hue='Free/Paid')
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.title('Pricing per subject/level')
sns.barplot(data=df5,x='subject',y='price',hue='level')
plt.show()
```
